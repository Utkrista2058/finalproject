```python
from joblib import load
model = load("final_model.pkl")
```

```python
from Modelfinal import y_test, y_test_predicted,X,y,X_train,X_test,y_train,y_test,best_dt_estimates,y_train_predicted,y_test_predicted
from sklearn.model_selection import train_test_split
```

```python

model = model
X_train = X_train
X_test = X_test
y_train = y_train
y_test = y_test


data_size_range = np.linspace(0.1, 1.0, 10) 


train_sizes = []
train_scores = []
test_sizes = []
test_scores = []


for data_size in data_size_range:
    n_samples = int(data_size * len(X_train))  
    X_train_sub, X_test_sub, y_train_sub, y_test_sub = train_test_split(X_train[:n_samples], y_train[:n_samples], test_size=0.2)  
    
    
    model.fit(X_train_sub, y_train_sub)
    
    
    train_score = model.score(X_train_sub, y_train_sub)
    test_score = model.score(X_test_sub, y_test_sub)
    train_sizes.append(n_samples)
    train_scores.append(train_score)
    test_sizes.append(len(X_test_sub))
    test_scores.append(test_score)


plt.plot(train_sizes, train_scores, label="Training Accuracy")
plt.plot(test_sizes, test_scores, label="Testing Accuracy")
plt.xlabel("Data Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve for Random Forest Classification")
plt.legend()
plt.show()
```
